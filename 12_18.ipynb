```python
import numpy as np
x=np.arange(3)
y=[100, 400, 900]

plt.bar(x, y, color=['r','g','b'],width=0.2)
plt.plot(x,y)

plt.xticks(x,['2018','2019','2020'])
#plt.yticks(x,['2018','2019','2020'])
plt.show()
```

```python
import numpy as np
y=np.arange(3)
x=[100, 400, 900]

plt.barh(y, x, color=['r','g','b'], height=0.2)
plt.plot(x,y)

plt.yticks(y, ['2018','2019','2020'])
#plt.yticks(x,['2018','2019','2020'])
plt.show()
```

```python
n=50

x=np.random.rand(n)
y=np.random.rand(n)

area=(30*np.random.rand(n))**2
colors=np.random.rand(n)

plt.scatter(x,y,area,colors,cmap='Spectral',alpha=0.5)
```

```python
ratio=[34,32,16,18]
labels=['Apple','Banana','Melon','Grape']
a={'width':0.7,'edgecolor':'k'}

plt.pie(ratio,labels=labels,autopct='%.1f%%', explode=[0,0.1,0,0.1],colors=['red','yellow','green','purple'],wedgeprops=a)
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

line1=plt.plot(months, sales_2019, label='2019')
line2=plt.plot(months, sales_2020, label='2020')
plt.xlabel('Month',loc='center')
plt.ylabel('Sales',loc='center')
plt.legend(ncol=1, fontsize=10, shadow=True)
plt.title('Monthly Sales Comparison (2019-2020)')
plt.grid(False,axis='both')
plt.show()
```

```python
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]


plt.bar(categories, data,width=0.7)
plt.title('Bar Chart')
plt.xlabel('Categories',loc='center')
plt.ylabel('Values',loc='center')
plt.grid(True,axis='both')
plt.ylim(0,50)
plt.xticks(rotation=45)
plt.show()
```

```python
sizes=[15,30,45,10]
labels=['Apple','Banana','Mango','Blueberry']
explode=(0,0.1,0,0)
a={'width':1,'edgecolor':'k'}

plt.pie(sizes,labels=labels,autopct='%.1f%%', explode=explode,colors=['blue','orange','green','red'], startangle=90, shadow=True)
plt.show()
```

```python
x = np.linspace(0, 10, 100)

# sin cos   
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.title('Sine and Cosine Functions')
plt.xlabel('X',loc='center')
plt.ylabel('Y',loc='center')
plt.grid(True,axis='both')

plt.plot(x,y_sin, label='sin(x)', color='blue')
plt.plot(x,y_cos, label='cos(x)',color='red')
plt.legend()
```
