```python
import pandas as pd
import folium

data=pd.read_csv('택시승차대 현황.csv', encoding='cp949')
wi=min(data.loc[:,'위도'])+(max(data.loc[:,'위도'])-min(data.loc[:,'위도']))/2
ky=min(data.loc[:,'경도'])+(max(data.loc[:,'경도'])-min(data.loc[:,'경도']))/2
m1=folium.Map(location=[wi,ky], zoom_start=15)

taxi = data.loc[:, ["위치명", "위도", "경도"]]
for index, row in taxi.iterrows():
    folium.Marker(location=[row[1], row[2]], popup='taxi',icon=folium.Icon(icon='fa-solid fa-taxi',prefix='fa-solid')).add_to(m1)
folium.Circle([wi,ky],radius=1000, color='#00ff00',fill_color='#ff0000').add_to(m1)


m1
```
